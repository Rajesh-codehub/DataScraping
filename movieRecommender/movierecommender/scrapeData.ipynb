```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
url1 = "https://www.imdb.com/list/ls048276758/"
movieNames = []
yearOfReleases= []
Ratings = []
plots = []
#certificates = []
Durations = []
genres = []
Directors = []

request = requests.get(url1)
if request.status_code == 200:
    soup = BeautifulSoup(request.text,"html.parser")
    movie_Data = soup.findAll('div',{"class":"lister-item mode-detail"})
    for data in movie_Data:
        name = data.h3.a.text
        year = data.h3.find('span', class_ = "lister-item-year text-muted unbold").text.strip("()")
        
        duration = data.p.find('span', class_ = 'runtime').text
        genre = data.p.find('span', class_ = 'genre').text
        rating = data.find('div', class_ = "ipl-rating-widget").find('div', class_ = "ipl-rating-star small").find('span', class_ = "ipl-rating-star__rating").text #find('div' class_ ''.find('div',class_ = "ipl-rating-star small")
        plot = data.find('div', class_ = "lister-item-content").find('p', class_ = '').text
        director = data.find('div', class_ = "lister-item-content").findAll('p', class_ = "text-muted text-small")[1].find('a').text
        # try:
        #     cert_data = data.p.find('span', class_ = 'certificate')
        #     certify = cert_data.text if cert_data else np.NaN
        # except AttributeError:
        #     certify = np.NaN

        movieNames.append(name)
        yearOfReleases.append(year)
        #certificates.append(certify)
        Durations.append(duration)
        genres.append(genre)
        Ratings.append(rating)
        plots.append(plot)
        Directors.append(director)
    print(len(movieNames),len(yearOfReleases),len(Durations),len(Ratings),len(plots),len(Directors),len(genres))

else:
    print("Request Failed")
```
